## Numpy array and basics 
  learn about numpy

```python
import numpy as np
```

### Creating array from list


```python
arr_1D=np.array([1,2,3,4,5])
print("1D array :",arr_1D)

arr_2D=np.array([[1,2,3],[4,5,6],[3,4,9]])
print("2D Array :",arr_2D)
```

### List vs numpy Array

```python
py_List=[1,2,3]
print("Multiplication of list :",py_List*2)

array=np.array([1,4,2])  #element wise multiplication
print("Multiplication of numpy array :",array*2)

import time
start=time.time()
py_List=[i*2 for i in range(100000)]
print("\n List operation time :",time.time()-start)

start=time.time()
array=np.arange(100000)*2
print("\n numpy operation time :",time.time()-start)
```

### Creating array from scratch

```python
zeros=np.zeros((3,4))     # creating zero element array of 3 rows and 4 columns
print("Zero Array :\n",zeros)

ones=np.ones((2,5))       # creating element one array of 2 rows and 5 columns
print("Ones Array :\n",ones)  

full=np.full((2,2), 8)        # creating constant value array by using full method np.full((rows,columns), const value)
print("Array of constant value 8:\n",full)

random=np.random.random((3,4))  # random values are in between 0 and 1
print("Random Array :\n",random)  

sequence = np.arange(0,11,2)  # creating seuence array by using arange(start,stop,steps) where stop are excluded
print("Sequence Array :\n",sequence)  
```

### Vector, Matrix and Tensor

```python
vector=np.array([1,2,3])
print("Vector :",vector)
# vector is 1D array

matrix = np.array([[1,2,3],
                   [4,5,6]])
print("Matrix :",matrix)
# matrix is 2D array

tensor=np.array([[[1,2,3],[4,5,6]],
                 [ [1,2,3],[9,3,4]],
                 [[0,6,7],[3,4,1]]])
print("Tensor :",tensor)
# tensor is n dimensional array
```

### Array Properties


```python
array=np.array([[1,2,3],[3,8,9]])
print("Shape :",array.shape)       # gives the detail of no. rows and columns
print("Dimension :",array.ndim)    # gives the dimension of the array 
print("Size :",array.size)         # gives the no. of total elements
print("type :",array.dtype)        # gives the datatype
```

### Array Reshaping

```python
arr=np.arange(15)
print("Original Array :",arr)

Reshaped=arr.reshape((3,5))
print("Reshaped Array :",Reshaped)

flattened=Reshaped.flatten()
print("Flattened Array :",flattened)

# ravels (return view ,instead of copy)
ravel=Reshaped.ravel()
print("Raveled Array :",ravel)

# Transpose
transpose=Reshaped.T
print("Transpose Array :",transpose)
```

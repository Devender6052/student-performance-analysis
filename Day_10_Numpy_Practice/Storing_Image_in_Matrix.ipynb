```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
array1=np.array([[1,2,3],[4,5,6]])
array2=np.random.rand(4,4)
array3=np.zeros((4,4))
print(array1)
print(array2)
print(array3)
```

```python
# Saving array in database
np.save('array1.npy',array1)
np.save('array2.npy',array2)
np.save('array3.npy',array3)
```

```python
# load the array
load_array1=np.load('array1.npy')
print(load_array1)
```

```python
try:
    logo=np.load('numpy-logo.npy')

    # display
    plt.figure(figsize=(10,6))
    plt.subplot(121)
    plt.imshow(logo)
    plt.title("Numpy Logo")
    plt.grid(False)

    dark_logo=1-logo

    plt.subplot(122)
    plt.imshow(dark_logo)
    plt.title("Dark logo")
    plt.grid(False)
except FileNotFoundError:
    print("Numpy logo not found")
```

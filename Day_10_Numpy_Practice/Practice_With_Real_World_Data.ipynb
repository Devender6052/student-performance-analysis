```python
import numpy as np
import matplotlib.pyplot as plt
```

```python

# Data structure: [restaurant_id, 2021, 2022, 2023, 2024]
sales_data = np.array([
    [1, 150000, 180000, 220000, 250000],  # Paradise Biryani
    [2, 120000, 140000, 160000, 190000],  # Beijing Bites
    [3, 200000, 230000, 260000, 300000],  # Pizza Hub
    [4, 180000, 210000, 240000, 270000],  # Burger Point
    [5, 160000, 185000, 205000, 230000]   # Chai Point
])

print("==== Zomato sales analysis ==== ")
print("Sales Data Shape :",sales_data.shape)
# print First Restaurant data
print("First Restaurant Data :",sales_data[0])
print("First 3 Restaurant Data :",sales_data[0:3])  #sales_data[:3]
print("collection data of all restaurant :",sales_data[:,1:])  #sales_data[:,1:len of column]
```

```python
# total sales per year
print(np.sum(sales_data,axis=0))

# total sales per year
yearly_total=np.sum(sales_data[:,1:],axis=0)
print(yearly_total)

# total sales per restaurant
res_total=np.sum(sales_data[:,1:],axis=1)
print(res_total)

# minimum sales per restaurant
min_sales=np.min(sales_data[:,1:],axis=1)
print(min_sales)

# maximum sales per year
max_sale=np.max(sales_data[:,1:],axis=0)
print(max_sale)

# Average sales per restaurant
avg_sale=np.average(sales_data[:,1:],axis=1)
# avg1_sale=np.mean(sales_data[:,1:],axis=1)
print(avg_sale)
```

```python
# cumulative sum
cumsum=np.cumsum(sales_data[:,1:],axis=1)
print(cumsum)

plt.Figure(figsize=(5,3))
plt.plot(np.mean(cumsum,axis=0))
plt.title("Average Cumulative sales Across all restaurants")
plt.xlabel("Years")
plt.ylabel("Sales")
plt.grid(True)
plt.show()
```

```python
vector1=np.array([1,2,3,4,5])
vector2=np.array([6,7,8,9,10])

print("Vector sum :",vector1+vector2)
print("Vector Multiplication :",vector1*vector2)
print("Dot product :",np.dot(vector1,vector2))

angle=np.arccos((np.dot(vector1,vector2))/(np.linalg.norm(vector1)*np.linalg.norm(vector2)))
print("angle :",angle)
```

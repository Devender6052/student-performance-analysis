```python
Emp_data = {
    "Name": ['Ram', 'Shyam', 'Ghanshyam', 'Dhanshyam', 'Aditi', 'Jagdish', 'Raj', 'Simran'],
    "Age": [28, 34, 22, 30, 29, 40, 25, 32],
    "Salary": [50000, 60000, 45000, 52000, 49000, 70000, 48000, 58000],
    "Performance Score": [85, 90, 78, 92, 88, 95, 80, 89]
}

emp_df=pd.DataFrame(Emp_data)
print("Descrptive details of the dataframe :")
print(emp_df.describe())

print("column names :",emp_df.columns)
print("Shape of the dataframe :",emp_df.shape)
```

## Filtering

```python
# Slecting secific column
column=emp_df["Age"]
print(column)

# selecting multiple columns
columns=emp_df[['Age','Salary']]
print(columns)

# selecting specific row under contition
rows =emp_df[emp_df['Age']>30]
print(rows)

# selecting rows under multiple conditions
rows1=emp_df[(emp_df["Salary"]>47000) & (emp_df["Salary"]<55000)]
print(rows1)
```
